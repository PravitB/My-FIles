# Linearly Seperable Data

```python
import numpy as np

# 1. Define the number of samples for each class
n_samples_per_class = 100

# 2. Create the first class's data points
# Class 0: centered around (0, 0)
X0 = np.random.randn(n_samples_per_class, 2) + np.array([0, 0])
y0 = np.zeros(n_samples_per_class)

# 3. Create the second class's data points
# Class 1: centered around (4, 4) to ensure linear separability
X1 = np.random.randn(n_samples_per_class, 2) + np.array([4, 4])
y1 = np.ones(n_samples_per_class)

# 4. Combine the data points from both classes into a single NumPy array (X)
X = np.vstack((X0, X1))

# 5. Create corresponding labels for each class and combine them into a single NumPy array (y)
y = np.hstack((y0, y1))

print(f"Shape of features (X): {X.shape}")
print(f"Shape of labels (y): {y.shape}")
print("First 5 rows of X:\n", X[:5])
print("First 5 labels of y:\n", y[:5])
print("Last 5 rows of X:\n", X[-5:])
print("Last 5 labels of y:\n", y[-5:])
```

```python
import matplotlib.pyplot as plt

# Plotting the linearly separable data
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1')
plt.title('Linearly Separable Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

# Non-Linearly Seperable Data

```python
import numpy as np

# 1. Define the number of samples for each class
n_samples_per_class_nl = 100

# 2. Create data points for the first class (inner circle)
# Class 0: points within a circle of radius 1 centered at (0, 0)
radius_inner = 1.0
r_inner = radius_inner * np.random.rand(n_samples_per_class_nl, 1)
theta_inner = 2 * np.pi * np.random.rand(n_samples_per_class_nl, 1)
X_inner = np.hstack((r_inner * np.cos(theta_inner), r_inner * np.sin(theta_inner)))
y_inner = np.zeros(n_samples_per_class_nl)

# 3. Create data points for the second class (outer annulus)
# Class 1: points between radius 2 and 3 centered at (0, 0)
radius_outer_min = 2.0
radius_outer_max = 3.0
r_outer = (radius_outer_min + (radius_outer_max - radius_outer_min) * np.random.rand(n_samples_per_class_nl, 1))
theta_outer = 2 * np.pi * np.random.rand(n_samples_per_class_nl, 1)
X_outer = np.hstack((r_outer * np.cos(theta_outer), r_outer * np.sin(theta_outer)))
y_outer = np.ones(n_samples_per_class_nl)

# 4. Combine the data points from both classes into a single NumPy array (X_non_linear)
X_non_linear = np.vstack((X_inner, X_outer))

# 5. Create corresponding labels for each class and combine them into a single NumPy array (y_non_linear)
y_non_linear = np.hstack((y_inner, y_outer))

print(f"Shape of non-linear features (X_non_linear): {X_non_linear.shape}")
print(f"Shape of non-linear labels (y_non_linear): {y_non_linear.shape}")
print("First 5 rows of X_non_linear:\n", X_non_linear[:5])
print("First 5 labels of y_non_linear:\n", y_non_linear[:5])
print("Last 5 rows of X_non_linear:\n", X_non_linear[-5:])
print("Last 5 labels of y_non_linear:\n", y_non_linear[-5:])
```

```python
import matplotlib.pyplot as plt

# Plotting the non-linearly separable data
plt.figure(figsize=(8, 6))
plt.scatter(X_non_linear[y_non_linear == 0][:, 0], X_non_linear[y_non_linear == 0][:, 1], color='green', label='Class 0 (Inner)')
plt.scatter(X_non_linear[y_non_linear == 1][:, 0], X_non_linear[y_non_linear == 1][:, 1], color='purple', label='Class 1 (Outer)')
plt.title('Non-Linearly Separable Data (Concentric Circles)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
